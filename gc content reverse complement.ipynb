```python
def find_reverse_complement(dna:str):
    #code dna into numbers
    dna = dna.replace("A", "1")
    dna = dna.replace("T", "2")
    dna = dna.replace("C", "3")
    dna = dna.replace("G", "4")
    #find complement
    complement = dna.replace("1","T").replace("2", "A").replace("3", "G").replace("4", "C")
    #reverse complement string and print
    return complement[::-1]





def find_gc_content(file_list: list):
    #create dict for pairing names with sequences
    sequence_dict = {}
    #add fasta names to gc_dict as keys with proceeding sequences as values

    key_name = ""
    for i in range(len(file_list)):
        if file_list[i].startswith(">"):
            key_name = file_list[i].strip()
            sequence_dict[key_name]= ""
        else:
            sequence_dict[key_name] += file_list[i].strip()
        
    #goes through each key in sequence_dict, prints key(sequence name)
    for key in sequence_dict:
        #print(key)
        #goes through each key value, finds the reverse complement, and prints
        print(key)
        print(find_reverse_complement(sequence_dict[key]))

f = open("rosalind_gc.txt")
test_list = f.readlines()

find_gc_content(test_list)

             
        
    
    
```
